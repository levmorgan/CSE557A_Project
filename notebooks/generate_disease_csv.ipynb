```python
corpus = {}
for index, row in df.iterrows():
    words = str(row['text']).split(' ')
    for word in words:
        if word in corpus:
            corpus[word] += 1
        else:
            corpus[word] = 1
```

```python
def keep_disease(words):
    #print(isinstance(words, float))
    if isinstance(words, float):
        words = str(words)
    else:
        words = [str(item) for item in words]
    return any(item in words for item in diseasewords)

df['keep'] = df.apply(lambda row : keep_disease(row['words']), axis = 1)
```

```python
import datetime

def create_timestamp(date):
    datestr = date.split(' ')
    d = datestr[0].split('/')
    t = datestr[1].split(':')
    #dt = datetime.datetime(int(d[2]), int(d[0]), int(d[1]), int(t[0]), int(t[1]), int(0))
    dt = datetime.datetime(int(d[2]), int(d[0]), int(d[1]))
    return dt

new_df['date'] = new_df.apply(lambda row : create_timestamp(row['Created_at']), axis = 1)
```

```python
new_df = new_df.sort_values(by=['date'])
```

```python
datecounts = {}
for index, row in new_df.iterrows():
    date = row['date'].strftime("%m/%d/%Y")
    if date in datecounts:
        datecounts[date] += 1
    else:
        datecounts[date] = 1
```
